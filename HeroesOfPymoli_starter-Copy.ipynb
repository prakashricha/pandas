```python
#Purchase Analysis(Age)
purchase_data1=pd.DataFrame(purchase_data)
```

```python
bins = [0,9,14,19,24, 29,34,39,100]

# Create the names for the four bins
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-40","40+"]
```

```python
purchase_data1["age demo"]=pd.cut(purchase_data1["Age"],bins,labels=group_names)
purchase_data
purchase_data1=purchase_data1.rename(columns={"Purchase ID":'purchaseid'})
purchase_data1 = purchase_data1.groupby(["age demo"])

purchasecount=purchase_data1.purchaseid.count()
purchasecount
totalpurchasevalue=purchase_data1.Price.sum()
agepurchase_df=pd.DataFrame(purchasecount)
agepurchase_df
avgpurchaseprice=round(totalpurchasevalue/purchasecount,2)
agepurchase_df=agepurchase_df.assign(averagepurchaseprice=avgpurchaseprice)
agepurchase_df=agepurchase_df.assign(totalvalpurchase=totalpurchasevalue)
temp=purchase_data1.SN.nunique()
temp
temp1=round(totalpurchasevalue/temp,2)
agepurchase_df=agepurchase_df.assign(ave_pricepurchase=temp1)
final_df=agepurchase_df.rename(columns={'purchaseid':'Purchase Count',
                                       'averagepurchseprice':'Average Purchase Price',
                                      'totalvalpurchase':'Total Purchase Value',
                                      'ave_pricepurchase':'Avg Total Purchase per Person' })
final_df
```

```python
# Most Popular Items
purchase_data1=pd.DataFrame(purchase_data)
```
